Now comes the mapping part:

```python
def continent_mapping(row):
    if row in Asia:
        return 'Asia'
    elif row in Africa:
        return 'Africa'
    elif row in NorthAmerica:
        return 'North America'
    elif row in SouthAmerica:
        return 'South America'
    elif row in Oceania:
        return 'Oceania'
    elif row in Europe:
        return 'Europe'
    else:
        return 'Unknown'

df['continent'] = df.country.apply(continent_mapping)
```

**We see that the total votes row is called '0', which can be fixed in a number of ways** 

1. transpose rows and columns, so that '0' now becomes a column, and changing a column name is pretty straightforward
2. first, second, third, 0 is an index column. Remove it as the index col, make it a normal row using reset_index, change '0' to 'total' and then again set it as the index column
3. We'll make a list of the index column, change 0 to total and then assign this list back to the index column.

we'll use the third approach

```python
final.index
```

```python
as_list = final.index.to_list()
as_list
```

```python
idx = as_list.index(0) # Meaning that pick the index value of the row with '0' as index
idx
```

```python
as_list[idx] = 'total' # basically as_list[3] = 'total'
as_list
```

```python
final.index = as_list
```

```python
final = final.sort_values(by='total',axis=1, ascending=False)
final #GREAT!
```

# GREAT WORK! NOW WE'LL GENERALIZE IT USING LOOPS and FUNCTIONS

```python
def looping(df):
    
    i = df['first'].value_counts()
    j = df['second'].value_counts()
    k = df['third'].value_counts() 
    
    player_wise = pd.DataFrame([i, j, k])
    total = pd.DataFrame(player_wise.sum())
    m = player_wise.append(total.transpose())
    as_list = m.index.to_list()
    idx = as_list.index(0)
    as_list[idx] = 'total'
    m.index = as_list
    m.fillna(0, inplace=True)
    m = m.astype(int)
    m = m.sort_values(by='total', axis=1, ascending=False)
    return m
    
```

```python
continent_dictionary = {}
continents = ['Asia','Africa','Europe','North America','South America','Oceania']
for x in continents:
    conti = df[df['continent'] == x]
    continent_dictionary[x] = looping(conti)
```
